```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/brain_MR.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Convert the image to grayscale for processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

### Implement Canny edge detection

```python
# Try Canny using "wide" and "tight" thresholds

wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 200, 240)
 
    
# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
```

```python
# Read in the image
image = cv2.imread('images/sunflower.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

## TODO: Define lower and upper thresholds for hysteresis
# right now the threshold is so small and low that it will pick up a lot of noise
lower_w = 40
upper_w = 120

lower_t = 120
upper_t = 240


wide_fl = cv2.Canny(gray, lower_w, upper_w)
tight_fl = cv2.Canny(gray, lower_t, upper_t)

# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide_fl, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight_fl, cmap='gray')
# plt.figure(figsize=(20,10))
# plt.imshow(edges, cmap='gray')
```

```python
# Canny with bigger Kernel size (default 3) formula
lower_w_k5 = 100
upper_w_k5 = 160

lower_t_k5 = 200
upper_t_k5 = 240

wide_fl_k5 = cv2.Canny(gray, lower_w_k5, upper_w_k5, apertureSize=5)
tight_fl_k5 = cv2.Canny(gray, lower_t_k5, upper_t_k5, apertureSize=5)

# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide - kernel 5')
ax1.imshow(wide_fl_k5, cmap='gray')

ax2.set_title('tight kernel 5')
ax2.imshow(tight_fl_k5, cmap='gray')
```

```python
# Canny with L2gradient formula (default False)
wide_fl_L2 = cv2.Canny(gray, lower_w, upper_w, L2gradient=True)
tight_fl_L2 = cv2.Canny(gray, lower_t, upper_t, L2gradient=True)

# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide - L2gradient')
ax1.imshow(wide_fl_L2, cmap='gray')

ax2.set_title('tight L2gradient')
ax2.imshow(tight_fl_L2, cmap='gray')
```
